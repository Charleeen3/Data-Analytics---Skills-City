```python
import pandas as pd
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/charlenehowarth/Desktop/Data Analytics - Skills City/Dataset-2.csv')

# Convert the 'Most_Recent_Revenue_USD' and 'Ransom_Amount' columns to floats
df['Most_Recent_Revenue_USD'] = df['Most_Recent_Revenue_USD'].apply(lambda x: float(x.replace('$', '').replace(',', '')))
df['Ransom_Amount'] = df['Ransom_Amount'].apply(lambda x: float(x.replace('$', '').replace(',', '')))

# Remove outliers using z-score
from scipy.stats import zscore
z_scores = zscore(df[['Most_Recent_Revenue_USD', 'Ransom_Amount']])
df = df[(z_scores < 3).all(axis=1)]

# Create a pair plot
sns.pairplot(df, height=3, aspect=1.5)
plt.suptitle('Pair Plot of Ransomware Attacks and Revenue', y=1.02)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/charlenehowarth/Desktop/Data Analytics - Skills City/Dataset-2.csv')

# Convert the 'Most_Recent_Revenue_USD' and 'Ransom_Amount' columns to floats
df['Most_Recent_Revenue_USD'] = df['Most_Recent_Revenue_USD'].apply(lambda x: float(x.replace('$', '').replace(',', '')))
df['Ransom_Amount'] = df['Ransom_Amount'].apply(lambda x: float(x.replace('$', '').replace(',', '')))

# Remove outliers using z-score
z_scores = zscore(df[['Most_Recent_Revenue_USD', 'Ransom_Amount']])
df = df[(z_scores < 3).all(axis=1)]

# Create a scatter plot with trend line and logarithmic scaling
plt.figure(figsize=(12, 8))
sns.regplot(data=df, x='Most_Recent_Revenue_USD', y='Ransom_Amount', scatter_kws={'s': 50, 'alpha': 0.5}, line_kws={'lw': 2})
plt.xscale('log')  # Logarithmic scaling on x-axis
plt.yscale('log')  # Logarithmic scaling on y-axis
plt.title('Ransomware Attacks and Revenue')
plt.xlabel('Most Recent Revenue (USD)')
plt.ylabel('Ransom Amount (USD)')
plt.grid(True, alpha=0.3)  # Add grid for better readability

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/charlenehowarth/Desktop/Data Analytics - Skills City/Dataset-2.csv')

# Convert 'Most_Recent_Revenue_USD' to numeric after removing dollar signs and commas
df['Most_Recent_Revenue_USD'] = pd.to_numeric(df['Most_Recent_Revenue_USD'].replace('[\\$,]', '', regex=True), errors='coerce')

# Drop rows with missing or NaN values in the 'Most_Recent_Revenue_USD' column
df = df.dropna(subset=['Most_Recent_Revenue_USD'])

# Create a box plot to compare Sub-Industry and revenue
plt.figure(figsize=(14, 8))
sns.boxplot(x='Sub-Industry', y='Most_Recent_Revenue_USD', data=df)
plt.yscale('log')  # Logarithmic scaling on y-axis for better visualization of wide range of revenue values
plt.title('Comparison of Sub-Industry and Revenue')
plt.xlabel('Sub-Industry')
plt.ylabel('Most Recent Revenue (USD)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```
